```python
import os
import pandas as pd
import librosa as lr
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
from sklearn.ensemble import RandomForestClassifier
import warnings 
warnings.filterwarnings(action= 'ignore')
```

```python
path = "D:\simple-recorder3\Final_Final1"
files = os.listdir(path)
df = pd.DataFrame(columns = ['max_pitch', 'avg_pitch', 'variance_pitch', 'harmonic',
       'harmonic_variance', 'percussive', 'percussive_variance', 'Chroma_cens',
       'Chroma_cens_var', 'chroma_cqt_mean', 'chroma_cqt_variance',
       'chroma_stft_mean', 'chroma_stft_var', 'mfcc_delta_mean',
       'mfcc_delta_variance', 'Mfccs', 'Mfccs_variance', 'Contrast',
       'Contrast_var', 'poly_features_mean', 'poly_features_variance',
       'Rolloff', 'Rolloff_var', 'Zrate', 'Zrate_var', 'Cent', 'Cent_var',
       'tonnetz_mean', 'tonnetz_var', 'spec_bw_mean', 'spec_bw_var',
       'rmse_mean', 'rmse_var', 'melspec_mean', 'melspec_var', 'max_lpc',
       'mean_lpc', 'var_lpc'])
l=0
for f in files: 
    file_name = path+'/'+f
    audio,sr_freq = lr.load(file_name)
    S = np.abs(lr.stft(audio))
    pitches,magnitudes = lr.core.piptrack(y = audio ,sr = sr_freq)
    
    max_pitch = np.max(pitches)
    avg_pitch = np.mean(pitches)
    variance_pitch = np.var(pitches)
                    
    y_harmonic, y_percussive = lr.effects.hpss(audio)
    harmonic = np.mean(y_harmonic)
    harmonic_variance = np.var(y_harmonic)
    percussive = np.mean(y_percussive)
    percussive_variance = np.var(y_percussive)
                    
    chroma=lr.feature.chroma_cens(y=y_harmonic, sr=sr_freq)
    Chroma_cens = np.mean(chroma)
    Chroma_cens_var = np.var(chroma)
    chroma_stft =lr.feature.chroma_stft(y=y_harmonic, sr=sr_freq)
    chroma_stft_mean = np.mean(chroma_stft)
    chroma_stft_var = np.var(chroma_stft)
                    
    chroma_cqt =lr.feature.chroma_cqt(y=y_harmonic, sr=sr_freq)
    chroma_cqt_mean = np.mean(chroma_cqt)
    chroma_cqt_variance = np.var(chroma_cqt)
                    
    poly_features = lr.feature.poly_features(S=S, sr=sr_freq)
    poly_features_mean = np.mean(poly_features)
    poly_features_variance = np.var(poly_features)
                    
    mfccs = lr.feature.mfcc(y=y_harmonic, sr=sr_freq)
    Mfccs = np.mean(mfccs)
    
    Mfccs_variance = np.var(mfccs)
    delta = lr.feature.delta(mfccs)
    mfcc_delta_mean = np.mean(delta)
    mfcc_delta_variance = np.var(delta)
                    
    melspectrogram = lr.feature.melspectrogram(y=audio, sr=sr_freq)
    melspec_mean = np.mean(melspectrogram)
    melspec_var = np.var(melspectrogram)
                    
    contrast=lr.feature.spectral_contrast(y=y_harmonic,sr=sr_freq)
    Contrast = np.mean(contrast)
    Contrast_var = np.var(contrast)
                    
    rolloff = lr.feature.spectral_rolloff(y=audio, sr=sr_freq)      
    Rolloff = np.mean(rolloff)
    Rolloff_var = np.var(rolloff)
                    
    zrate=lr.feature.zero_crossing_rate(y_harmonic)
    Zrate = np.mean(zrate)
    Zrate_var = np.var(zrate)
                    
    cent = lr.feature.spectral_centroid(y=audio, sr=sr_freq)
    Cent = np.mean(cent)
    Cent_var = np.var(cent)
                    
    tonnetz = lr.feature.tonnetz(y=audio, sr=sr_freq)
    tonnetz_mean = np.mean(tonnetz)
    tonnetz_var = np.var(tonnetz)
                    
    spec_bw = lr.feature.spectral_bandwidth(y=audio, sr=sr_freq)
    spec_bw_mean = np.mean(spec_bw)
    spec_bw_var = np.var(spec_bw)
                                         
    rmse=lr.feature.rms(y=audio)[0]
    rmse_mean = np.mean(rmse)
    rmse_var = np.var(rmse)
                                         
    melspectrogram = lr.feature.melspectrogram(y=audio, sr=sr_freq)
    melspec_mean = np.mean(melspectrogram)
    melspec_var = np.var(melspectrogram)
    
    lpc=lr.lpc(audio, order=16)
    max_lpc=np.max(lpc)
    mean_lpc=np.mean(lpc)
    var_lpc=np.var(lpc)
    
    data = list([max_pitch, avg_pitch,variance_pitch,harmonic,harmonic_variance,percussive, percussive_variance, Chroma_cens,Chroma_cens_var, chroma_cqt_mean, chroma_cqt_variance,chroma_stft_mean, chroma_stft_var, mfcc_delta_mean,mfcc_delta_variance, Mfccs, Mfccs_variance, Contrast,Contrast_var, poly_features_mean, poly_features_variance,Rolloff, Rolloff_var, Zrate, Zrate_var, Cent, Cent_var,tonnetz_mean,tonnetz_var,spec_bw_mean,spec_bw_var,rmse_mean, rmse_var, melspec_mean, melspec_var, max_lpc,mean_lpc,var_lpc])
    df.loc[l] = data
    l+=1
```

```python
df.to_csv("Team_features.csv",index=False)
```

```python
x = df.drop(['Target'] , axis = 1).values
y = df['Target'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=0.2 , random_state = 42)
```

```python
ro_scaler = RobustScaler()
x_train = ro_scaler.fit_transform(x_train)
x_test = ro_scaler.fit_transform(x_test)
```

```python
from sklearn.svm import SVC
svm = SVC(gamma='auto')
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, precision_score, f1_score,confusion_matrix, mean_absolute_error , r2_score , mean_squared_error, mean_absolute_percentage_error
def resultsSummarizer(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro')
    rec = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')
    sensitivity = recall_score(y_true,y_pred,average='macro')
    
    
    print(f'Accuracy Score: ' + '{:.4%}'.format(acc))
    print(f'Precision Score: ' + '{:.4%}'.format(prec))
    print(f'Recall Score: ' + '{:.4%}'.format(rec))
    print(f'F_1 Score: ' + '{:.4%}'.format(f1))
    print(f'Sensitvity: ' + '{:.4%}'.format(sensitivity))
```
